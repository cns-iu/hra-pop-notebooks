## Load corridor names

```python
folder_path = "../../../hra-pop/output-data/v0.10.2/corridors/"
corridor_ids = set()

# Loop through files in the corridors folder to get names (IDs) and add them to a list
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    if os.path.isfile(file_path):
      file_name = os.path.splitext(os.path.basename(file_path))[0]
      
      # add the file name to the set
      corridor_ids.add(file_name)  # Outputs: filename
        

pprint(corridor_ids)
```
