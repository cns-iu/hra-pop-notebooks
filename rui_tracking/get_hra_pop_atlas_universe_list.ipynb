## A note book to get HRAPop ATLAS and UNIVERSE datasets a list

```python
# install and import libraries
%pip install pandas

import pandas as pd
```

```python
# load data
hra_pop_version = 'v0.11.1'
url = f'https://raw.githubusercontent.com/x-atlas-consortia/hra-pop/refs/heads/main/output-data/{hra_pop_version}/reports/universe-ad-hoc/sankey.csv'

data = pd.read_csv(url)
data
```

### Atlas

```python
# get atlas only
df_atlas = data[data['is_atlas_dataset'] == True].drop_duplicates(
    ['unique_dataset_id']).drop(columns=['is_azimuth_reference', 'cell_type_annotation_tool'])
df_atlas
```

```python
# export to csv
df_atlas.to_csv('output/atlas.csv', index=False)
```

### Universe

```python
# get universe only
df_universe = data.drop_duplicates(
    ['unique_dataset_id']).drop(columns=['is_azimuth_reference', 'cell_type_annotation_tool'])
df_universe
```

```python
# export to csv
df_universe.to_csv('output/universe.csv', index=False)
```
